```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_csv("https://api.covid19india.org/csv/latest/raw_data.csv")
```

```python
raw_data['Date Announced'] = pd.to_datetime(raw_data['Date Announced'],format = '%d/%m/%Y')
# ,format = '%d/%m/%Y'
```

```python
raw_data.head(2)
```

## Initial set of patients

India got the first case of corona virus on 30-jan-2020, reported in Kerala. First patient is known to be female of age 20, a student from Wuhan, China. Currently she has recovered. Here's the district wise list of cases in the first month 

```python
city_data = raw_data.sort_values(['Date Announced']).head(200).groupby('Detected State')[['Patient Number']].count().sort_values(by ='Patient Number', ascending = False)
city_data
```

From the table above, we can gather that Gurugram had the highest number of covid cases for the first 200 cases reeported in India.

This can be due to the fact that :-
1. Gurugram is financial and technology hub where many people might have come from business trips
2. Gurugram has  ~17,48,000 companies enablling more contacts which would have resulted in more cases
3. Gurugram is upper middle class area , hence we have lot of students returning from abroad.



```python
state_data = raw_data.sort_values(['Date Announced']).head(200).groupby('Detected City')[['Patient Number']].count().sort_values(by ='Patient Number', ascending = False)
state_data
```

Maharashtra has the highest no. of cases from first 200 cases reported in  India.

We want to see which patient has been responsible for other cases

```python
raw_data.groupby(['Contracted from which Patient (Suspected)'])[['Patient Number']].count().sort_values(by='Patient Number', ascending = False)
```

```python
raw_data.query(" `Contracted from which Patient (Suspected)` ==  'E0' ").head(10)
```

From the above datasets, it seems that, Patient E0 has affected 958 patients. Further it has been idenfitified as E0 patient has Indonesian Nationality and thus it can be considered that patients belongs to Tablighi Jamat Conference held in Delhi.

Importing new datasets Statewise daily covid data and statewise tests data


```python
statewise = pd.read_csv("https://api.covid19india.org/csv/latest/state_wise_daily.csv")
statewise_test = pd.read_csv("https://api.covid19india.org/csv/latest/statewise_tested_numbers_data.csv")
```

```python
statewise.head()
```

This data contains daily count of confirmed,recovered and deceased cases. Column TT seems to be a sum of the group counts hence can be removed.

```python
statewise = statewise.drop(columns = 'TT')
```

Changing shape of statewise data for better visualization

```python
s1 = statewise.melt(id_vars=['Date','Status']).sort_values(['Date','Status'])
#statewise.melt(id_vars=['Date','Status']).sort_values(['Date','Status','variable'])
s1
```
